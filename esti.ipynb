```python
# reading raw data file
columns = ['idiot_num','class','time', 'ax', 'ay', 'az']
df_har = pd.read_csv('WISDM_ar_v1.1_raw.txt', header = None, names = columns)
df_har.head(10)
```

```python
# removing null values
df_har = df_har.dropna()
df_har.shape
# drop rows where timestamp is 0
df = df_har[df_har['time'] != 0]
# arrange data in ascending order of user and timestamp
df = df.sort_values(by = ['idiot_num', 'time'], ignore_index=True)
```

```python
activities = ['Walking', 'Jogging', 'Upstairs', 'Downstairs', 'Sitting', 'Standing']
for i in activities:
    data = df[(df['idiot_num'] == 36) & (df['class'] == i)][:400]
    sns.lineplot(y = 'ax', x = 'time', data = data)
    sns.lineplot(y = 'ay', x = 'time', data = data)
    sns.lineplot(y = 'az', x = 'time', data = data)
    plt.legend(['ax','ay','az'])
    plt.ylabel(i)
    plt.title(i, fontsize = 15)
    plt.show()
```

```python
random_seed = 42   
n_time_steps = 50 
n_features = 3 
step = 10 
n_classes = 6 
n_epochs = 50       
batch_size = 1024   
learning_rate = 0.0025
l2_loss = 0.0015
```

```python
segments = []
labels = []
sig=[]
for i in range(0,  df.shape[0]- n_time_steps, step):  

    xs = df['ax'].values[i: i + 50]

    ys = df['ay'].values[i: i + 50]

    zs = df['az'].values[i: i + 50]

    label = stats.mode(df['class'][i: i + 50])[0][0]
    xss=re.findall("\d+\.\d+",str(xs))
    yss=re.findall("\d+\.\d+",str(ys))
    zss=re.findall("\d+\.\d+",str(zs))
  
  
    reshaped_segments = np.asarray(yss, dtype= np.float32)
    f_min=1
    f_max=20
    X_Dat=[]
    cur_t=np.linspace(0, 1, reshaped_segments.shape[0])

    my_data = np.squeeze(reshaped_segments)
    print("______________Dim_______________")
    print(my_data.shape)
    
    cur_x=my_data
    scalogram = cwt(cur_x, 1/reshaped_segments.shape[0], 2, f_min, f_max)
    fig=plt.figure()
    ax = fig.add_subplot(111)
    x, y = np.meshgrid(cur_t,
    np.logspace(np.log10(f_min), np.log10(f_max), scalogram.shape[0]))

    ax.pcolormesh(x, y, np.abs(scalogram), cmap=obspy_sequential)
    fig.savefig("tep.png")
    image = cv2.imread("tep.png")# My reference Image
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    arr=gray.flatten()
    farr=np.expand_dims(arr, axis=0)
    sig.append(farr)
    print("______________New Dim_______________")
    print(farr.shape)

    labels.append(label)


labels = np.asarray(pd.get_dummies(labels), dtype = np.float32)
```

```python
sig = np.asarray(sig, dtype = np.float32)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sig, labels, test_size = 0.2, random_state = random_seed)
```

```python
model = keras.Sequential()
model.add(layers.GRU(128, input_shape=(1,8000)))
model.add(layers.Dense(6, activation = 'softmax'))
# Dropout layer
model.add(layers.Dropout(0.5)) 
# Dense layer with ReLu
model.add(layers.Dense(units = 64, activation='relu'))
# Softmax layer
model.add(layers.Dense(y_train.shape[1], activation = 'softmax'))
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs = n_epochs, validation_split = 0.20, batch_size = batch_size, verbose = 1)
```

```python
plt.plot(np.array(history.history['loss']), "r--", label = "Train loss")
plt.plot(np.array(history.history['accuracy']), "g--", label = "Train accuracy")
plt.plot(np.array(history.history['val_loss']), "r-", label = "Validation loss")
plt.plot(np.array(history.history['val_accuracy']), "g-", label = "Validation accuracy")
plt.title("Training session's progress over iterations")
plt.legend(loc='lower left')
plt.ylabel('Training Progress (Loss/Accuracy)')
plt.xlabel('Training Epoch')
plt.ylim(0) 
plt.show()
```
